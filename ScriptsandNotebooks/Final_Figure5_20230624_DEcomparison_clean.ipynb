```r
for (i in 1:8){
    tmp <- gsub("fSupplT7_DE_SejongIn_", "", get(paste0("DEfile", i)), perl=TRUE)
    tmp5 <- gsub("0vs1_FullROI_N_cellselection_Y_", "", tmp, perl=TRUE)
    tmp2 <- gsub("_ROI_large_Vent_LV_DsSelect", "", tmp5)
    tmp3 <- sub("_[NY]_OL.*", "", tmp2, perl=TRUE)
    tmp6 <- gsub("_Non", "", tmp3, perl=TRUE)
    tmp7 <- gsub("_NES_HCMP", "", tmp6, perl=TRUE)
    tmp4 <-get(paste0("DE", i))
    tmp4 <- tmp4 %>% mutate(COI = tmp7) 
    assign(paste0("DE", i), tmp4)
    print(tmp7)
}
```

```r
for (i in 1:8){
    tmp4 <-get(paste0("DE", i))
    tmp4 <- tmp4 %>% mutate(Index = 1:nrow(tmp4)) 
    assign(paste0("DE", i), tmp4)
}
```

```r
for (i in 1:8){
    tmp4 <-get(paste0("DE", i))
    tmp4$PosIndex[tmp4$logFC > 0 & tmp4$adj.P.Val < 0.05]<-1:nrow(tmp4[tmp4$logFC > 0 & tmp4$adj.P.Val < 0.05,])
    tmp4$NegIndex[tmp4$logFC < 0 & tmp4$adj.P.Val < 0.05]<-1:nrow(tmp4[tmp4$logFC < 0 & tmp4$adj.P.Val < 0.05,])
    assign(paste0("DE", i), tmp4)
}
```

```r
# DE_Noseg<-rbind(DE1, DE2, DE3)
# DE_eachcell<-rbind(DE4, DE5, DE6)
# DE_othercell<-rbind(DE7, DE8, DE9)
DE_total1 <- rbind(DE1, DE2, DE3, DE4, DE5, DE6, DE7, DE8)
# DE_total2 <- rbind(DE11, DE12, DE13, DE14, DE15, DE16)
```

```r
unique(DE_total1$COI )
```

```r
DE_total1$diffexpressed <- "NO"
DE_total1$diffexpressed[DE_total1$logFC > 0 & DE_total1$adj.P.Val < 0.05] <- "UP"
DE_total1$diffexpressed[DE_total1$logFC < 0 & DE_total1$adj.P.Val < 0.05] <- "DOWN"
DE_total1$delabel <- NA
DE_total1$delabel[DE_total1$PosIndex %in% c(1:5) | DE_total1$NegIndex %in% c(1:5)] <- DE_total1$SYMBOL[DE_total1$PosIndex %in% c(1:5) | DE_total1$NegIndex %in% c(1:5)]
DE_total1$COI <- factor(DE_total1$COI, levels=c(
    "P_Hyp_Cardiomyocytes",
     "P_Deg_Cardiomyocytes",
     "P_Fib_Cardiomyocytes",
     "P_Disarray_Cardiomyocytes",
     "P_Hyp_Endothelial_cells",
     "P_Deg_Endothelial_cells",
     "P_Fib_Endothelial_cells",
     "P_Disarray_Endothelial_cells")
)
```

```r
DE_total1$COI <- recode_factor(DE_total1$COI, 
     P_Hyp_Cardiomyocytes = "CMC - Hypertrophy",
     P_Deg_Cardiomyocytes = "CMC - Degeneration",
     P_Fib_Cardiomyocytes = "CMC - Fibrosis",
     P_Disarray_Cardiomyocytes = "CMC - Disarray",
     P_Hyp_Endothelial_cells = "EC - Hypertrophy",
     P_Deg_Endothelial_cells = "EC - Degeneration",
     P_Fib_Endothelial_cells = "EC - Fibrosis",
     P_Disarray_Endothelial_cells = "EC - Diarray"
                               ) 
                               
                               
```

```r
options(repr.plot.width=18, repr.plot.height=5)
```

```r
ggplot(data=DE_total1, aes(x=COI, y=(-log10(adj.P.Val))*(sign(logFC)), col=diffexpressed, label=delabel)) +
geom_jitter(width=0.02) +
theme_minimal() +
geom_text_repel(size=5, max.overlaps=100
) +
scale_color_manual(values=c("blue", "black", "red"), labels=c('Down-regulated', 'No change', 'Up-regulated')) +
geom_hline(yintercept=c(-log10(0.05), log10(0.05)), col="red") +
geom_hline(yintercept=0, col="black") +
theme(axis.text.x=element_text(angle = 10, hjust=0.5, size=14,face="bold"),
        axis.title=element_text(size=14,face="bold"),
     legend.title=element_blank(),
     legend.text = element_text(size=12),
     ) +
xlab("") +
guides(color = guide_legend(override.aes = list(size = 8))) 
```
